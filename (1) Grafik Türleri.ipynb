# Veri Görüntüleme

Verilerle çevrili bir dünyada, onları anlamak, yaşamanın gerekliliklerinden biridir. Görselleştirme bu noktada yardımımıza koşarak herhangi bir miktardaki veriyi anlamamıza yardımcı olur.

- Veri görselleştirme, verilerin grafiksel olarak tasvir edilmesiyle ilgili bir çalışma alanıdır.

- Veri görselleştirme, bilgileri verimli bir şekilde iletmek için istatistiksel grafikler, çizelgeler ve diğer teknikleri kullanır; sayısal veriler ise noktalar, çizgiler ve çubuklar kullanılarak kodlanabilir.

- Görselleştirme kavramlarının çoğu, 1900'lerin ortalarında başlayan bilgisayar çağından önce bile bulunmuştu.

- Sonuç olarak: Veri görselleştirmenin bilgi işlemle hiçbir ilgisi yoktur, ancak verileri sunma ve işleme yollarıyla bilgisayarlara bağlıdır.

- Verileri doğru şekilde görüntülemek çok önemlidir; bu nedenle yılları sırasız olarak sıralamaktan, gereksiz renkler kullanmaktan veya gerekmedikçe tabloya 0'dan başlamamaktan kaçınmak en iyisidir.

## Grafik Bütünlüğü için Tufte'nin 6 İlkesi

Tufte, tasarımın altı temel ilkesini önerir: karşılaştırmaları göster, nedenselliği göster, çok değişkenli verileri kullan, modları (metin, resimler, sayılar gibi) tamamen entegre et, güvenilirlik oluştur ve içeriğe odaklan. Her biri, belirli bir veri gösterimi için tanımladığınız hedefleri tamamen kucaklamaya yönelik olmalıdır. Her ilke için, görselleştirmelerinizi geliştirmek üzere nasıl uygulanacağına ilişkin örnekleri özetliyoruz.

1. Karşılaştırmalar: Bağımlı değişkenler arasındaki zıtlıkları ve farklılıkları göstermek için verileri karşılaştırmalarla (çubuk grafikler ve benzeri) gösterin.

2. Nedensellik: Bir veya daha fazla bağımsız değişkenin bağımlı değişkenleri nasıl etkilediğini veya etkilediğini gösterin.

3. Çok değişkenlilik: Çeşitli veriler birleştirilir, böylece bir izleyici başka türlü karmaşık bir anlatıyı kolayca yorumlayabilir.

4. Entegrasyon: Kaynak verilerden bulgulara kanıt göstermek için çeşitli bilgi modlarından (metinler, haritalar, hesaplamalar, diyagramlar, vb.) yararlanın.

5. Dokümantasyon: Güvenilirlik için atıf, ayrıntılı başlıklar ve ölçümleri (ölçekler) dahil edin.

6. Bağlam: Öncesi ve sonrası durumunu tanımlayın veya tasvir edin. Gelecekteki sonuçlara dair ipucu vermek için trend çizgilerini gösterin.

# Verileri Alma ve Araştırma

**Bu bölümde** elmaslar hakkında bilgi içeren Diamonds veri setini kullanacağım.

Veri kümesinde 10 sütun vardır:
- **karat:** pırlantanın ağırlığı (0,2--5,01)
- **kesim:** kesim kalitesi (Orta, İyi, Çok İyi, Birinci Sınıf, İdeal)

- **renk:** elmas rengi, J'den (en kötü) D'ye (en iyi)
- **berraklık:** elmasın ne kadar net olduğunun ölçümü (I1 (en kötü), SI2, SI1, VS2, VS1, VVS2, VVS1, IF (en iyi))
- **derinlik:** toplam derinlik yüzdesi = z / ortalama(x, y) = 2 * z / (x + y) (43--79)
- **tablo:** baklava tepesinin en geniş noktaya göre genişliği (43--95)
- **fiyat:** ABD doları cinsinden fiyat (\326--\18.823$)
- **x:** mm cinsinden uzunluk (0--10,74)
- **y:** genişlik (mm) (0--58,9)
- **z:** mm cinsinden derinlik (0--31,8)

# Grafik Türleri

## Çubuk Grafikler

- Değerlerin nasıl dağıldığını görsel olarak araştırmak için kümelenmiş sayısal değerlerin histogramlarını çizmek ve ayrı bir öğe grubunun miktar dalgalanmasını göstermek için kullanılır.

### Yatay Çubuk Çizimi

```python
# "barh()" yönteminin sonundaki "h" harfi "yatay" anlamına gelir
df["cut"].value_counts().plot.barh().set_title("Class Frequencies of the Cut Variable")
```

### Dikey Çubuk Grafiği

```python
df["cut"].value_counts().plot.bar().set_title("Class Frequencies of the Cut Variable")
```

```python
sns.barplot(x = "cut", y = df.cut.index, data = df, color = "#af0d9d").set_title("Cut Density")
```

### Gruplanmış Çubuk Grafik

- Amaç, çubuk grafiği oluşturan değişkenlerden birinin alt gruplarını göstermek olduğunda kullanılır.

- Başka bir değişkenin ton eşlemesi ile koşullu dağılımları çizer

- Renkler ve sıralamaları her zaman birbiriyle uyumludur.

```python
sns.barplot(x = "cut",
            y = "price",
            hue = "color",
            data = df)
plt.title("Cut - Price - Color")
```

```python
sns.barplot(x = "cut",
            y = "price",
            hue = "clarity",
            data = df)
plt.title("Cut - Price - Clarity")
```

## Pasta Grafiği

```python
df.info()
```

```python
df.groupby("color").sum().plot.pie(y = "price", autopct = "%.1f%%", ylabel="", legend = False)
```

## Dağılım grafiği

```python
cols = ["x", "y", "z"]
for col in cols:
    sns.scatterplot(x = col, y = "price", data = df, color = "#5be1f3")
    plt.show()
```

## Histogram

- Histogram, verilerin parametre olarak verilen bölme sayısına göre bölümler halinde sınıflandırıldığı bir diyagramdır.

- Seaborn kitaplığının histplot() işlevi, aksi belirtilmediği sürece kutu sayısını otomatik olarak atar.

```python
sns.histplot(df["price"], kde = True, color = "#953319", bins=50)
```

## Joint Plot

- Bir ortak grafik, varsayılan olarak, verilen değişkenlerin 2 histogramından ve bunların korelasyonunu temsil eden 1 dağılım grafiğinden oluşur.

- Parametreleri değiştirerek, dağılım grafiği yerine KDE grafiği de kullanılabilir.

```python
sns.jointplot(x = "price", y = "carat", data =df, color = "#6c35fa")
```

## Catplot

- Kategorik Grafik, kategorik bir değişkenin öğelerine karşılık gelen sayısal bir değişkenin saçılımını görselleştirmek için kullanılır.

```python
sns.catplot(x = "cut", y = "price", data = df)
plt.title("Cut - Price")
```

```python
sns.catplot(x = "cut",
            y = "price",
            hue = "color",
            kind = "point",
            data = df)
```

### KDE Grafiği (Çekirdek Yoğunluğu Tahmini)

- KDE Grafiği, sürekli bir değişkenin Olasılık Yoğunluğunu görselleştirmek için kullanılır

- Tek bir grafikte birden çok çizim de görüntülenebilir

```python
sns.kdeplot(df["price"], color = "#05d814")
```

```python
sns.FacetGrid(df, # İlk satırda FacetGrid fonksiyonu kullanılarak bir ızgara (grid) oluşturulur.
              hue = "cut",
              height=5, # height parametresi, grafiğin yüksekliğini belirler. 
              xlim=(0,10000)) # xlim parametresi ise x ekseninin sınırlarını belirler.
.map(sns.kdeplot, "price", shade=True).add_legend()
```

### Korelasyon Matrisi

- Bir korelasyon ısı haritası, iki sütun arasında bir 2B korelasyon matrisini (tablo) göstermek için renkli hücreler kullanır

- Korelasyon açısından 1, iki sütun arasında güçlü bir pozitif ilişkiyi, -1 ise aralarında güçlü bir negatif ilişkiyi temsil eder.

- İki değişkenli analizin hangi sütunlar üzerinde yapılacağı korelasyon matrisinde verilen değerlere göre seçilmesi daha doğru bir yöntemdir.

```python
sns.heatmap(df.corr(), annot=True)
```
